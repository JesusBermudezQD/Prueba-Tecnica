```python
import pandas as pd
import numpy as np
from datetime import datetime
```

### GenerateCategoryData
##### Esta función cumple la tarea de generar cierta cantidad de categorías según el valor que se le envíe en el parámetro. Inicia tomando una lista predefinida con nombres de categorías. Luego, mediante un ciclo, se va concatenando de manera aleatoria cualquier elemento de la lista con un valor numérico. Este valor numérico permite la creación de categorías únicas que se asignan a una lista. Después, se genera una lista de IDs desde el 1 hasta el valor que llega como parámetro. Posteriormente, crea un DataFrame con estas listas y lo retorna.

```python
def GenerateCategoryData(cant):
    cant+=1
    name=["Technology","Bussiness","Art","Science","Health"]
    category_list=[]
    count=1
    while count!=cant:
        data=f"{np.random.choice(name)} {np.random.randint(1,cant)}"
        if( data not in category_list):
            #print(count)
            category_list.append(data)
            count=count+1
        
    ids=list(range(1,cant))
    return pd.DataFrame({"id":ids,"name":category_list})
category_data=GenerateCategoryData(13)
category_data
```

### GenerateLevelData
##### Esta función cumple la tarea de generar cierta cantidad de niveles según el valor que se le envíe en el parámetro. Inicia tomando una lista predefinida con nombres de categorías. Luego, mediante un ciclo, se va concatenando de manera aleatoria cualquier elemento de la lista con un valor numérico. Este valor numérico permite la creación de categorías únicas que se asignan a una lista. Después, se genera una lista de IDs desde el 1 hasta el valor que llega como parámetro. Posteriormente, crea un DataFrame con estas listas y lo retorna.

```python
def GenerateLevelData(cant):
    cant+=1
    name=["Beginner","Intermediante","Advanced"]
    category_list=[]
    count=1
    while count!=cant:
        date=f"{np.random.choice(name)} {np.random.randint(1,cant)}"
        if( date not in category_list):
            #print(count)
            category_list.append(date)
            count=count+1
        
    ids=list(range(1,cant))
    return pd.DataFrame({"id":ids,"name":category_list})
level_data=GenerateLevelData(13)
level_data
```

### GenerateCursesData
##### Esta función cumple la tarea de generar cierta cantidad de cursos según el valor que se le envíe en el parámetro. retorna un dataframe el cual contiene las siguientes columnas:
- id: rango generado desde el 1 hasta el numero enviado como parametro.
- name: nombre del curso generado apartir de una cadana de texto seguido por un numero que va desde el 1 hasta el numero enviado por parametro mas 1 ya que de esta manera nos aseguramos que los nombres lleguen a la cantidad esperada.
-  category_id: generada apartir de la columna id del dataframe resultante de ejecutar la funcion GenerateCategoryData y se toman al azar la cantidad de valores que determine el valor enviado por parametro
- level_id: generada apartir de la columna id del dataframe resultante de ejecutar la funcion GenerateLevelData y se toman al azar la cantidad de valores que determine el valor enviado por parametro
- start_date: generada apartir de la funcion datetime y la funcion random para crear fechas aleatorias para la cantidad de registros que determine el valor enviado por parametro

```python
def GenerateCursesData(cant):
    
    return pd.DataFrame({
        "id":range(1,cant+1),
        "name":[f"Course {i}" for i in range(1,cant+1)],
        "category_id":np.random.choice(list(category_data["id"]),cant),
        "level_id":np.random.choice(list(level_data["id"]),cant),
        "start_date": [datetime(2021,np.random.randint(1,13),np.random.randint(1,29)) for _ in range(cant)]
    })

curses_data=GenerateCursesData(100)
```
